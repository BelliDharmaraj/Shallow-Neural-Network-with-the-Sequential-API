```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=df, x="mpg")
plt.title("Title of the chart")
plt.show()
```

```python
sns.histplot(data=df, x="mpg", bins=50).set(title="Title of Chart", ylabel="Title of Y-axis")
plt.show()
```

```python
sns.boxplot(data=df, y="horsepower")
plt.title("Boxplot for horsepower")
plt.show()
```

```python
plt.scatter(data=df, x="weight", y="displacement")
plt.title("displacement with respect to weight")
plt.ylabel("displacement")
plt.xlabel("weight")
plt.show()
```
